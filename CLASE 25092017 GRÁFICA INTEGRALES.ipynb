```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
def walker():
    global x,y
    r = random.uniform(0,1)
    if r< 0.01: #un 1% de probabilidad de tomar un paso grande
        x += np.random.normal(0,200)
        y += random.gauss(0,200)
    else:
        x += np.random.normal(0,1)
        y += random.gauss(0,1)
    return [x,y]


x = 0 #Inicia el caminador
y = 0

Pos = np.zeros((2,1000))

for i in np.arange(0,len(Pos[0]),1):
    Pos[:,i] = walker()
    
plt.plot(Pos[0,:],Pos[1,:],'bo-')
plt.plot(Pos[0,-1],Pos[1,-1],'ro-')
```

```python
inicio = np.array([ [0], [0] ])

elements = np.array( [ [1,-1,0,0], [0,0,1,-1] ] )
probabilites = [0.4,0.2,0.2,0.2]

pasos = elements[:,np.random.choice(elements.shape[1], 1000, p =probabilites)]
pasos = np.concatenate( (inicio,pasos), axis = 1 )

Pos = np.cumsum (pasos ,axis = 1)
plt.plot(Pos[0,:], Pos[1,:], 'bo-')
```

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
# Numero de puntos tirados aleatoriamente
N = 1000
f = lambda x: np.sin(x)
a,b = 0, np.pi
Altura = 1

x = np.random.uniform(a,b,N)
y = np.random.uniform(a,Altura,N)
X = np.arange(a,b,.01)
fx = f(x)

plt.plot(x[y>fx], y[y>fx], "go")
plt.plot(x[y<fx], y[y<fx], "ro")
plt.plot(X,f(X))

N_0 = len(x[y<fx])
Base = b-a
yest = N_0/N*(Base*Altura)
print(yest)
```
