```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
%matplotlib inline
import matplotlib as mpl
label_size = 14
mpl.rcParams[ 'xtick.labelsize'] = label_size
mpl.rcParams[ 'ytick.labelsize'] = label_size
```

```python
k = 3.0 # Constante de muelle
m = 1.0 # Masa
B = 0.5 # Constante de amortiguación

def armonico(variables, t):
    x, y = variables
    return [y, -k * x / m - B / m * y]

inicial = [0.6, 0.4] # Vector de posición inicial y velocidad inicial
# condiciones inciciales  x(t=0)=0.6 [m] y (t=0)=0.4 [m/s]

#tiempo = mp.linspace(0, 15) #
tiempo = np.arange(0, 20, 0.1)

resultados = odeint(armonico, inicial, tiempo)
```

```python
xx, yy = resultados.T #extraer posicion y velocidad.
```

```python
plt.plot(tiempo, xx, c = 'r', label = "Posición")
plt.plot(tiempo, yy, c = 'k', label = "Velocidad")
plt.legend(loc = 'best', prop={'size': 14})
plt.xlabel('tiempo', fontsize = 14)
plt.show()
```

```python
plt.plot(xx, yy, c = 'r', label = "Posición")
plt.legend(loc = 'best', prop={'size': 14})
plt.xlabel('tiempo', fontsize = 14)
plt.show()
```
